```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import rand
```

```python
x = np.linspace (-(2*np.pi), 2*np.pi, 50)
y= np.sin(x) 
plt.plot(x, y, marker = "h", color = 'violet')
plt.title("Sin Function")
plt.grid()
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```
